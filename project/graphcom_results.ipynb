```python
import matplotlib.pyplot as plt

# Example: Multiple accuracy values for different models or parameters
model_names = ['AlexNet','ResNet18', 'Random Forest Classifier', 'K-Nearest Neighbors']
accuracy_values = [55.87,68.25,63.00,67.00]
colors = ['skyblue', 'skyblue', 'lightgreen','lightgreen']

# Plotting the accuracy values
plt.figure(figsize=(10, 6))
bars = plt.bar(model_names, accuracy_values, color=colors)
plt.xlabel('Deeep learning and Machine Learning Classification Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparision of Deeep learning and Machine Learning Classification Models')
plt.ylim(0, 100)  # Set the y-axis limit to represent accuracy values between 0 and 1

# Adding legend
legend_labels = ['AlexNet','ResNet18', 'Random Forest Classifier', 'K-Nearest Neighbors']
plt.legend(bars, legend_labels)

plt.show()
```

# Importing required libraries

```python
import warnings
warnings.filterwarnings('ignore')
import os
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV
```

# Loading, Preprocessing, Normalizing, Data augmentaion and Feature extraction

```python
def load_and_normalize_image(path):
    img = cv2.imread(path)
    if img is None:
        raise ValueError(f"Image cannot be read. Check the file format and path: {path}")
    img = cv2.resize(img, (256, 256))
    img = img.astype(np.float32) / 255.0
    return img

def extract_hog_features(image):
    features, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16),
                              cells_per_block=(1, 1), visualize=True, channel_axis=-1)
    return features

def augment_image(image):
    rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
    flipped = cv2.flip(image, 1)
    return rotated, flipped

def prepare_dataset(base_directory):
    features = []
    labels = []
    supported_formats = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif')  # Add or remove formats as needed

    # Debugging to check the path
    print(f"Base directory: {base_directory}")
    print(f"Directories in base directory: {os.listdir(base_directory)}")

    for label_dir in os.listdir(base_directory):
        class_dir = os.path.join(base_directory, label_dir)
        if os.path.isdir(class_dir):  # Check if it is a directory
            print(f"Processing folder: {label_dir}")
            for filename in os.listdir(class_dir):
                if filename.lower().endswith(supported_formats):  # Checking against supported formats
                    path = os.path.join(class_dir, filename)
                    print(f"Processing file: {filename}")
                    try:
                        img = load_and_normalize_image(path)
                        img_features = extract_hog_features(img)
                        features.append(img_features)
                        labels.append(label_dir)  # Using folder name as label

                        # Data augmentation
                        rotated, flipped = augment_image(img)
                        features.append(extract_hog_features(rotated))
                        labels.append(label_dir)
                        features.append(extract_hog_features(flipped))
                        labels.append(label_dir)
                    except Exception as e:
                        print(f"Error processing {filename}: {e}")
        else:
            print(f"Skipped: {label_dir} is not a directory")

    return np.array(features), labels
```

# Encode labels as integers

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

# Hyperparameter Tuning and  Random forest Model Training

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

params = {
    'max_depth': [2,3,5,10,20],
    'max_features': ['auto', 'sqrt'],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200]
}

# Instantiate the grid search model
grid_search_rf= GridSearchCV(estimator=rf,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")

# Fit GridSearchCV to the training data
grid_search_rf.fit(X_train, y_train_encoded)
```

# Evaluate Random Forest

```python
rf_best = grid_search_rf.best_estimator_
rf_predictions = rf_best.predict(X_test)
```

```python
print("Random Forest Classification Report:")
print(classification_report(y_test_encoded, rf_predictions))
```

# Hyperparameter Tuning and  K-Nearest neighbors Model Training

```python
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['ball_tree', 'kd_tree', 'brute', 'auto'],
    'leaf_size': [10, 30, 50],
    'metric': ['minkowski', 'euclidean', 'manhattan'],
    'p': [1, 2]
}
```

```python
# Grid search with cross-validation for KNN
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=3, verbose=2, n_jobs=-1)
grid_search_knn.fit(X_train, y_train_encoded)
```

# Evaluate KNN model

```python
knn_best = grid_search_knn.best_estimator_
knn_predictions = knn_best.predict(X_test)
```
